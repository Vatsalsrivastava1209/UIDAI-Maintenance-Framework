```python
def load_and_concat_csvs(folder_path, data_label):
    """
    Checks if a folder exists, reads all CSVs inside, 
    and returns a single concatenated DataFrame.
    """
    if os.path.exists(folder_path):
        print(f"‚úÖ Folder '{folder_path}' found.")
        path_pattern = os.path.join(folder_path, "*.csv")
        files = glob.glob(path_pattern)
        
        if files:
            try:
                # Load all files in the folder
                df_list = [pd.read_csv(f) for f in files]
                df_final = pd.concat(df_list, ignore_index=True)
                print(f"üéâ Success {data_label}! Shape: {df_final.shape}")
                return df_final
            except Exception as e:
                print(f"‚ùå Error reading {data_label} CSVs: {e}")
                return pd.DataFrame()
        else:
            print(f"‚ö†Ô∏è Folder found, but NO csv files in '{folder_path}'")
            return pd.DataFrame()
    else:
        print(f"‚ùå Folder '{folder_path}' NOT found.")
        return pd.DataFrame()

# --- Execution Section ---

base_dir = "datasets"

# 2. Define folder names (joined with the base directory)
folder_enrol = os.path.join(base_dir, "api_data_aadhar_enrolment")
folder_bio   = os.path.join(base_dir, "api_data_aadhar_biometric")
folder_demo  = os.path.join(base_dir, "api_data_aadhar_demographic")

# # 1. Define folder names
# folder_enrol = "api_data_aadhar_enrolment"
# folder_bio   = "api_data_aadhar_biometric"
# folder_demo  = "api_data_aadhar_demographic"

# 2. Call the function for each dataset
df_master  = load_and_concat_csvs(folder_enrol, "Enrolment")
df_master1 = load_and_concat_csvs(folder_bio, "Biometric")
df_master2 = load_and_concat_csvs(folder_demo, "Demographic")

# 3. Final Summary Check
print("\n" + "="*30)
if not df_master.empty and not df_master1.empty and not df_master2.empty:
    print("üöÄ ALL DATAFRAMES READY FOR ANALYSIS!")
else:
    print("‚ö†Ô∏è Warning: One or more DataFrames are empty. Check folder paths.")
```

```python
# 1. Define the cleanup map
state_map = {
    'WEST  BENGAL': 'WEST BENGAL',
    'WEST BANGAL': 'WEST BENGAL',
    'WESTBENGAL': 'WEST BENGAL',
    'ORISSA': 'ODISHA',
    'PONDICHERRY': 'PUDUCHERRY',
    'JAMMU & KASHMIR': 'JAMMU AND KASHMIR',
    'ANDAMAN & NICOBAR ISLANDS': 'ANDAMAN AND NICOBAR ISLANDS',
    'DADRA & NAGAR HAVELI': 'DADRA AND NAGAR HAVELI AND DAMAN AND DIU',
    'DADRA AND NAGAR HAVELI': 'DADRA AND NAGAR HAVELI AND DAMAN AND DIU',
    'DAMAN AND DIU': 'DADRA AND NAGAR HAVELI AND DAMAN AND DIU',
    'DAMAN & DIU': 'DADRA AND NAGAR HAVELI AND DAMAN AND DIU',
    'THE DADRA AND NAGAR HAVELI AND DAMAN AND DIU': 'DADRA AND NAGAR HAVELI AND DAMAN AND DIU'
}

# 2. Create a function to clean any dataframe
def finalize_states(df):
    # Standardize string format first
    df['state'] = df['state'].astype(str).str.strip().str.upper()
    
    # Apply the mapping
    df['state'] = df['state'].replace(state_map)
    
    # Remove the numeric '100000' anomaly
    df = df[df['state'] != '100000']
    
    return df

# 3. Apply to all your datasets
df_master = finalize_states(df_master)
df_master1 = finalize_states(df_master1)
df_master2 = finalize_states(df_master2)
```

## Update Rate
Update Rate is the ratio of Aadhaar updates to total enrolments in a given state and time period.

Lower values may indicate identity maintenance backlog or disengagement.

```python
enrol_state = df_master.groupby("state")[['age_0_5', 'age_5_17', 'age_18_greater']].sum().sum(axis=1).to_frame("total_enrolments")
upd_state = df_master2.groupby("state")[['demo_age_5_17', 'demo_age_17_']].sum().sum(axis=1).to_frame("updates")

engagement = enrol_state.join(upd_state, how="inner").fillna(0)

# Robust Update Rate calculation (Fix 2)
engagement["update_rate"] = (
    engagement["updates"] / engagement["total_enrolments"]
).replace([np.inf, -np.inf], np.nan).fillna(0)
```

## Update Lag

```python

# 2. ANALYZE: Create the state_summary for Update Lag
state_summary = (
    engagement
    .groupby("state")
    .agg(
        total_enrolments=("total_enrolments", "sum"),
        avg_update_rate=("update_rate", "mean")
    )
    .reset_index()
)

# 3. FLAG: Apply the binary risk signal
state_summary["update_lag_flag"] = np.where(
    (state_summary["total_enrolments"] > state_summary["total_enrolments"].median()) &
    (state_summary["avg_update_rate"] < state_summary["avg_update_rate"].median()),
    "High Risk",
    "Normal"
)

# 4. VIEW: Show the High Risk states
print("States Flagged as High Risk (High Volume, Low Update Activity):")

print(state_summary[state_summary["update_lag_flag"] == "High Risk"])
```

```python
# Updated dictionary to include the city-to-state mappings
extra_fixes = {
    # Spelling Fixes
    'UTTARANCHAL': 'UTTARAKHAND',
    'CHHATISGARH': 'CHHATTISGARH',
    'WEST BENGLI': 'WEST BENGAL',
    
    # City/District to State Mappings
    'JAIPUR': 'RAJASTHAN',
    'NAGPUR': 'MAHARASHTRA',
    'DARBHANGA': 'BIHAR',
    'MADANAPALLE': 'ANDHRA PRADESH',
    'BALANAGAR': 'TELANGANA',
    'PUTTENAHALLI': 'KARNATAKA',
    'RAJA ANNAMALAI PURAM': 'TAMIL NADU'
}

# Apply to your dataframes
df_master['state'] = df_master['state'].replace(extra_fixes)
df_master1['state'] = df_master1['state'].replace(extra_fixes)
df_master2['state'] = df_master2['state'].replace(extra_fixes)

# Now, when you run your variance/balance code, 
# these rows will merge into the correct 36 states!
```

```python
demo_long = df_master2.melt(
    id_vars=['state'], 
    value_vars=['demo_age_5_17', 'demo_age_17_'], 
    var_name='age_group', 
    value_name='value_count' # Changed this name to avoid the conflict
)

# 2. Calculate Totals per State/Age Group
demo_rates = demo_long.groupby(["state", "age_group"])["value_count"].sum().reset_index()

# 3. Calculate Variance per State
variance = demo_rates.groupby("state")["value_count"].var().fillna(0)

# 4. Calculate the Balance Score
demo_balance = (1 / (1 + variance)).to_frame("balance_score")
balance = 1 / (1 + variance)

# 5. Calculate the Total Updates per State (The "Permanent" Column logic)
# Now we create the state-level totals to join with our balance score
state_totals = demo_rates.groupby("state")["value_count"].sum().to_frame("total_updates")

# 6. Final Merge
final_analysis = state_totals.join(demo_balance)

print(final_analysis.sort_values(by="balance_score", ascending=False).head(10))
```

```python
# 1. Re-build the engagement table
enrol_state = df_master.groupby("state")[['age_0_5', 'age_5_17', 'age_18_greater']].sum().sum(axis=1).to_frame("enrolments")
upd_state = df_master2.groupby("state")[['demo_age_5_17', 'demo_age_17_']].sum().sum(axis=1).to_frame("updates")
engagement = enrol_state.join(upd_state, how="inner").fillna(0)
engagement["update_rate"] = engagement["updates"] / engagement["enrolments"]

# 2. Calculate Demographic Balance (Consistency)
demo_long = df_master2.melt(id_vars=['state'], value_vars=['demo_age_5_17', 'demo_age_17_'])
variance = demo_long.groupby(["state", "variable"])['value'].sum().groupby('state').var().fillna(0)
balance = (1 / (1 + variance)).to_frame("balance_score")

# 3. Normalize all components (0 to 1)
scaler = MinMaxScaler()
engagement["enrol_norm"] = scaler.fit_transform(engagement[["enrolments"]])
engagement["update_rate_norm"] = scaler.fit_transform(engagement[["update_rate"]])
engagement["balance_norm"] = scaler.fit_transform(balance[["balance_score"]])

# 4. Final Inclusion Index
index_df = engagement[engagement['enrolments'] > 1000].copy()
index_df["inclusion_index"] = (
    0.40 * index_df["enrol_norm"] + 
    0.35 * index_df["update_rate_norm"] + 
    0.25 * index_df["balance_norm"]
)

index_df = index_df.sort_values("inclusion_index", ascending=False)
print(index_df[["enrolments", "inclusion_index"]].head(10))
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# 1. Initialize Scaler
scaler = MinMaxScaler()

# 2. Normalize the Update Rate
# This translates the raw rates into a 0 to 1 scale
engagement["update_rate_norm"] = scaler.fit_transform(
    engagement[["update_rate"]]
)

# 3. Normalize the Balance Score 
# We use the 'balance_score' column created in your previous step
engagement["balance_norm"] = scaler.fit_transform(balance[["balance_score"]])

# 4. Create a Final "Aadhaar Performance Index"
# Applying the 50/50 weights for activity and equity
engagement["final_score"] = (engagement["update_rate_norm"] + engagement["balance_norm"]) / 2

# 5. See the Top 10 States
print("Aadhaar Performance Index: Top 10 States")
print(engagement[['update_rate_norm', 'balance_norm', 'final_score']]
      .sort_values(by="final_score", ascending=False)
      .head(10))
```

```python
# 1. RE-BUILD: Establishing the base engagement table
# Using standardized column names for total enrolments and updates
enrol_state = df_master.groupby("state")[['age_0_5', 'age_5_17', 'age_18_greater']].sum().sum(axis=1).to_frame("total_enrolments")
upd_state = df_master2.groupby("state")[['demo_age_5_17', 'demo_age_17_']].sum().sum(axis=1).to_frame("updates")

engagement = enrol_state.join(upd_state, how="inner").fillna(0)

# 2. ROBUST CALCULATION: Handling potential division by zero
engagement["update_rate"] = (
    engagement["updates"] / engagement["total_enrolments"]
).replace([np.inf, -np.inf], np.nan).fillna(0)

# 3. NORMALIZATION: Scaling components from 0 to 1
scaler = MinMaxScaler()
engagement["enrol_norm"] = scaler.fit_transform(engagement[["total_enrolments"]])
engagement["update_rate_norm"] = scaler.fit_transform(engagement[["update_rate"]])

# Use the 'balance_score' column from your demographic variance calculation
# We scale it here so it's ready for the index
engagement["balance_norm"] = scaler.fit_transform(balance[["balance_score"]])

# 4. RISK ASSESSMENT: Filtering and Indexing
# We keep states with > 1000 enrolments for statistical significance
index_df = engagement[engagement['total_enrolments'] > 1000].copy()

# Re-framing as Identity Maintenance Risk
# High population = High Risk | Low Updates = High Risk | Low Balance = High Risk
index_df["identity_maintenance_risk"] = (
    (0.40 * (1 - index_df["update_rate_norm"])) + 
    (0.35 * (1 - index_df["balance_norm"])) + 
    (0.25 * index_df["enrol_norm"])
)

# 5. FINAL CLASSIFICATION: Sorting and Tiers
index_df = index_df.sort_values("identity_maintenance_risk", ascending=False)

index_df["risk_level"] = pd.qcut(
    index_df["identity_maintenance_risk"],
    q=3,
    labels=["Low Risk", "Medium Risk", "High Risk"]
)

# 6. DISPLAY: The Policy Leaderboard
print("Top 10 States Requiring Identity Maintenance Outreach:")
print(index_df[["total_enrolments", "identity_maintenance_risk", "risk_level"]].head(10))
```

```python
# 1. RE-ASSIGN: Syncing with your latest data
index_df = engagement[engagement['total_enrolments'] > 1000].copy()

# 2. ENSURE COLUMN EXISTS: Calculate the Risk Index right here to avoid KeyError
# We use the 'Risk' logic: High population (norm) and Low updates (1-norm)
index_df["identity_maintenance_risk"] = (
    0.40 * (1 - index_df["update_rate_norm"]) +   # Fix: Low update = High risk
    0.35 * (1 - index_df["balance_norm"]) +      # Fix: Low balance = High risk
    0.25 * index_df["enrol_norm"]                 # Fix: High population = High impact
)

# 3. REDEFINE: The robust function
def compute_index(df, w_enrol, w_update, w_balance):
    """Calculates a weighted index using normalized components."""
    return (
        w_enrol  * df["enrol_norm"] + 
        w_update * df["update_rate_norm"] + 
        w_balance * df["balance_norm"]
    )

# 4. CALCULATE: The alternative index
index_df["index_alt"] = compute_index(index_df, 0.35, 0.40, 0.25)

# 5. RANKING: Now 'identity_maintenance_risk' is guaranteed to exist
index_df["rank_original"] = index_df["identity_maintenance_risk"].rank(ascending=False, method='min')
index_df["rank_alt"]      = index_df["index_alt"].rank(ascending=False, method='min')

# 6. MEASURE: Rank Shift
index_df["rank_shift"] = index_df["rank_original"] - index_df["rank_alt"]

# 7. RESULT
print("Sensitivity Analysis: Top 10 States by Rank Shift")
print(index_df[[
    "rank_original", 
    "rank_alt", 
    "rank_shift"
]].sort_values("rank_shift", ascending=False).head(10))
```

```python
# 1. Define Policy Intervention Status
# We initialize all states as 'Stable'
index_df["intervention_priority"] = "Monitor"

# 2. Identify the 'Update Gap' (High Impact, Low Maintenance)
# Mask: High Enrolment (top 30%) AND Low Update Rate (bottom 30%)
mask = (index_df["enrol_norm"] > 0.7) & (index_df["update_rate_norm"] < 0.3)

index_df.loc[mask, "intervention_priority"] = "IMMEDIATE OUTREACH"

# 3. View the Target List for UIDAI
# We sort by the Risk Index to see the most critical states first
priority_list = index_df[index_df["intervention_priority"] == "IMMEDIATE OUTREACH"]

print("UIDAI Priority List: States with Critical Update Backlogs")
print(priority_list[[
    "total_enrolments", 
    "update_rate", 
    "identity_maintenance_risk", 
    "intervention_priority"
]].sort_values("identity_maintenance_risk", ascending=False))
```
